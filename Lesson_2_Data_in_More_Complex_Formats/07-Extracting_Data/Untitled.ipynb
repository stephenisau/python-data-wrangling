```python
import xml.etree.ElementTree as ET #import library

article_file = "exampleResearchArticle.xml" #

def get_root(filename):
    tree = ET.parse(filename)
    return tree.getroot()
get_root(article_file)
```

```python
tree = ET.parse(article_file)
root = tree.getroot()
root
```

```python
DATACOUNT = []
tagCount = {}

for child in root.iter():
    if child.tag not in tagCount:
        tagCount[child.tag] = 1
    else:
        tagCount[child.tag] += 1
DATACOUNT.append(tagCount)
print(DATACOUNT)
```
