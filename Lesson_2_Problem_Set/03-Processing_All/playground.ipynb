extract the flight data from that table as a list of
dictionaries, each dictionary containing relevant data from the file and table
row. This is an example of the data structure you should return:

data = [{"courier": "FL",
         "airport": "ATL",
         "year": 2012,
         "month": 12,
         "flights": {"domestic": 100,
                     "international": 100}
        },
         {"courier": "..."}
]

```python
data = []
info = dict()
info['carrier'] = 'temp_courier'
info['airport'] = 'temp_airport'
#print(info)
for tr in trs:
    if not 'TOTAL' in str(tr):
        trdict = dict()
        trdict['carrier'] = info['carrier']
        trdict['airport'] = info['airport']
        trdict['year'] = int(tr.findAll('td')[0].text)
        trdict['month'] = int(tr.findAll('td')[1].text)
        flightdict = dict()
        flightdict['domestic'] = int(tr.findAll('td')[2].text.replace(',',''))
        flightdict['international'] = int(tr.findAll('td')[3].text.replace(',',''))
        trdict['flights'] = flightdict
        data.append(trdict)
print(data)
```
