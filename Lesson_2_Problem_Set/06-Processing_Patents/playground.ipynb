```python
import xml.etree.ElementTree as ET
PATENTS = 'patent.data'

tree = ET.parse(PATENTS)
tree.getroot()
```

```python
with open(PATENTS, 'r') as reader:
    print(type(reader))
    for line in reader:
        print(line)
        print(type(line))
```

```python
with open(PATENTS, 'r') as reader:
    for i in enumerate(reader):
        print(i)
```

```python
with open(PATENTS, 'r') as fin:
    originalfile = fin.readlines()
    print(len(originalfile))
    print(originalfile)
    
```

```python
with open(PATENTS, 'r') as fin:
    filenum = -1
    writer = None
    for line in fin:
        if line.find('<?xml version="1.0" encoding="UTF-8"?>') != -1:
            print(line)
            if writer:
                writer.close()
            filenum += 1
            writer = open("{}-{}.data".format("patent", filenum), 'w')
            print(filenum)
            print(writer)
        writer.write(line.strip())
        print(writer.write(line.strip()))
    writer.close()
    
                
            
```
