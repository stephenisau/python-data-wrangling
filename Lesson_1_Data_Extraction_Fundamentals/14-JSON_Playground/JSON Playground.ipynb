```python
def query_site(url, params, uid="", fmt="json"): #4 parameters
    """
    This is the main function for making queries to the musicbrainz API. The
    query should return a json document.
    """
    params["fmt"] = fmt #set params format to json
    r = requests.get(url + uid, params=params) #send a get request to URL
    print ("requesting: "+ str(r.url))#print message while sending request

    if r.status_code == requests.codes.ok:#if status_code matches request code
        return r.json()#return json object
    else:
        r.raise_for_status()
#if not, return the status?
```

```python
def query_by_name(url, params, name):# 3 parameters
    """
    This adds an artist name to the query parameters before making an API call
    to the function above.
    """
    params["query"] = "artist:" + name
    return query_site(url, params)
query_by_name(ARTIST_URL, query_type['simple'], 'Nirvana')
```

```python
def pretty_print(data, indent=4):
    """
    After we get our output, we can use this function to format it to be more
    readable.
    """
    if type(data) == dict:#if data type is dictionary,
        print (json.dumps(data, indent=indent, sort_keys=True))
        #this prints out the json object as a string, sorts the keys
    else:
        print (data)
        #if not then this happens
```
