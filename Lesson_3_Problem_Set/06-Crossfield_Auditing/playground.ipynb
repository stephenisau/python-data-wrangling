```python
import csv
import pprint

CITIES = 'cities.csv'
FIELDS = ["point", "wgs84_pos#lat", "wgs84_pos#long"]

DATA = []

with open(CITIES, 'r') as f:
    reader = csv.DictReader(f)
    header = reader.fieldnames
    for i in range(3):
        next(reader)
    for line in reader:
        fieldValues = {}
        for field in FIELDS:
            fieldValues[field] = line[field]
        DATA.append(fieldValues)
        
        
DATA
        
```
